```python
pd.crosstab(df['Airline'],df['Source'],normalize='columns')*100


#Maximum jet planes are leaving from Kolkata,Bangalore,Delhi,Kolkata,Mumbai
#Maximum indigo planes are leaving from Chennai
#People from delhi mostly prefer to go through multiple carriers
```

```python
pd.crosstab(df['Airline'],df['Total_Stops'],normalize='columns')*100

#  Most of the indigo flights are non stop
#  Only Air India flights have 4 stops
#  Air India have Maximum 3 stops and two stops
#  Jet Airways have maximum 1 stops
```
